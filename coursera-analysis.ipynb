```python
import pandas as pd
import plotly.express as px


# Plotting the sunburst chart
fig = px.sunburst(clean_main_data, path=['Keyword', 'Offered By'], values='Number of Review', 
                  color='Keyword', title='Reviews distribution by "category of course" and the "institution offering the course"',
                  color_continuous_scale='RdBu', width=1000, height=1000)
fig.show()
```
